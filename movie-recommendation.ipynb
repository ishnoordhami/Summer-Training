```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df=pd.read_csv(r"C:\Users\dhami\Downloads\archive (19)\ml-latest-small\movies.csv")
df
```

```python
df.drop(columns=['movieId'],inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df["genres"] = df["genres"].str.replace("|", " ", regex=False)
```

```python
df
```

```python
tfidf = TfidfVectorizer()
```

```python
tfidf_matrix = tfidf.fit_transform(df["genres"])
```

```python
def recommend(movie_title, top_n=10):
    movie_idx = df[df["title"].str.lower() == movie_title.lower()].index
    if len(movie_idx) == 0:
        print("Movie not found")
        return
    movie_idx = movie_idx[0]
    selected_movie = tfidf_matrix[movie_idx]
    scores = (tfidf_matrix @ selected_movie.T).toarray().flatten()
    indices = scores.argsort()[::-1]
    print(f"\nMovies similar to '{movie_title}':\n")
    count = 0
    for idx in indices:
        if idx != movie_idx:
            print(df.iloc[idx]["title"])
            count += 1
        if count == top_n:
            break
```

```python
recommend("Toy Story (1995)")
```
