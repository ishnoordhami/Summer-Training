## Loan Approval Prediction

```python
import pandas as pd
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

## Data Loading

```python
df=pd.read_csv(r"C:\Users\dhami\Downloads\archive (29)\loan_data.csv")
df
```

## Data Cleaning

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
lb=LabelEncoder()
df['person_gender']=lb.fit_transform(df['person_gender'])
df['person_education']=lb.fit_transform(df['person_education'])
df['person_home_ownership']=lb.fit_transform(df['person_home_ownership'])
df['loan_intent']=lb.fit_transform(df['loan_intent'])
df['previous_loan_defaults_on_file']=lb.fit_transform(df['previous_loan_defaults_on_file'])
```

```python
df
```

## K Nearest Neighbors Classifier

```python
x=df.drop('loan_status',axis=1)
y=df[['loan_status']]
```

```python
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
clf=KNeighborsClassifier(n_neighbors=7)
```

```python
clf.fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

```python
y_pred=clf.predict(x_test)
y_pred
```

```python
class_report=classification_report(y_test,y_pred)
print(class_report)
```
