```python
#create a plot with 9 subplots with x label, y label and title, label on bar, gap between the plots
#1- line
#2- stack
#3- box plot
#4- histogram
#5- bar chart
#6- violin chart
#7- horizontal bar chart
#8- scatter plot
#9- area chart

x=[1,2,3,4,5]
ax1=plt.subplot(331)
plt.xlabel("no. of points")
plt.ylabel("data values")
plt.title("line plot")
ax1.plot(x)

x=[1,2,3,4,5]
y=[1,4,3,2,1]
ax1=plt.subplot(332)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("area plot")
plt.stackplot(x,y)

x=[1,2,3,4,5]
y=[[1,4,3,2,1],[3,2,3,1,0],[4,2,1,4,4]]
ax1=plt.subplot(333)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("stackplot chart")
plt.stackplot(x,y)

x=[1,2,3,4,510,102,30,40,33,44]
ax1=plt.subplot(334)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("box plot")
plt.boxplot(x)

x=[1,2,3,4,10,14,5,15,8,6,6,12,10,18,20,7,5,1,0]
ax1=plt.subplot(335)
plt.xlabel("bins")
plt.ylabel("frequency")
plt.title("histogram")
plt.hist(x, bins=5)

x=[1,2,3,4]
y=[10,20,30,40]
ax1=plt.subplot(336)
plt.xlabel("index")
plt.ylabel("data values")
plt.title("bar graph")
ax1.bar(x,y)

x=[1,2,3,4,510,102,30,40,33,44]
ax1= plt.subplot(337)
plt.xlabel("no. of points")
plt.ylabel("data values")
plt.title("violin chart")
ax1.violinplot(x)

x=[1,2,3,4]
y=[10,20,30,40]
ax1=plt.subplot(338)
plt.xlabel("index")
plt.ylabel("data values")
plt.title("bar graph")
ax1.barh(x,y)

x=[1,2,3,4,5]
y=[10,20,30,40,50]
ax1=plt.subplot(339)
plt.xlabel("no. of points")
plt.ylabel("data values")
plt.title("scatter plot")
plt.scatter(x,y)

plt.subplots_adjust(wspace=0.5, hspace=1)

plt.show()
```

```python
# pie cahrt (used for percentage)
sizes=[10,20,30,40]
labels=['a','b','c','d']
explode=(0,0,0,0) #cutting or highlighting
ax1=plt.subplot()
ax1.pie(sizes,explode = explode, labels = labels, autopct="%1.2f%%")#1-whole no
plt.show()
```

```python
sizes=[10,20,30,40]
labels=['a','b','c','d']
explode=(0.1,0,0,0) #cutting or highlighting
ax1=plt.subplot()
ax1.pie(sizes,explode = explode, labels = labels)
plt.show()
```
