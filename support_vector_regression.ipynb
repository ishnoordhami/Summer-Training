# Used Car Price Prediction

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.svm import SVR
from sklearn.metrics import r2_score, mean_squared_error
```

## Data Loading

```python
df = pd.read_csv(r"C:\Users\dhami\Downloads\archive (27)\car data.csv")
```

```python
df
```

## Data Cleaning

```python
df.duplicated().sum()
```

```python
df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
lb=LabelEncoder()
df['Car_Name']=lb.fit_transform(df['Car_Name'])
df['Fuel_Type']=lb.fit_transform(df['Fuel_Type'])
df['Seller_Type']=lb.fit_transform(df['Seller_Type'])
df['Transmission']=lb.fit_transform(df['Transmission'])
```

```python
df
```

## Support Vector Regression

```python
x=df.drop('Selling_Price',axis=1)
y=df[['Selling_Price']]
```

```python
x_train,x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
svr = SVR(kernel='rbf')
```

```python
svr.fit(x_train,y_train)
```

```python
y_pred = svr.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Squared Error:", mse)
print("R2 Score:", r2)
```
